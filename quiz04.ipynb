### Question 1: Create the `nobel_laureates` table with the following columns:
- `id` (INTEGER, PRIMARY KEY, AUTOINCREMENT)

- `year_award` (INTEGER, NOT NULL)
- `prize_amount` (REAL, NOT NULL)
- `age_at_award` (INTEGER, NOT NULL)

```python
# Write your answer for Question 1 here.
# Hint: After creating the table, you can verify it exists with:
# cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name='nobel_laureates';")
# print("Table created successfully!") if cursor.fetchone() else print("Table not found!")

query1 = """
CREATE TABLE noble_laureates (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    field TEXT NOT NULL,
    year_award INTEGER NOT NULL,
    prize_amount REAL NOT NULL,
    age_at_award INTEGER NOT NULL
);
"""

cursor.execute(query1)
print("Table 'noble_laureates' created successfully.")
```

### Question 2: Create the `discoveries` table with the following columns:
- `id` (INTEGER, PRIMARY KEY, AUTOINCREMENT)
- `laureate_id` (INTEGER, FOREIGN KEY referencing nobel_laureates(id), with ON DELETE CASCADE)
- `discovery` (TEXT, NOT NULL)
- `year_discovery` (INTEGER, NOT NULL)
- `research_funding` (REAL, NOT NULL)
- `citation_count` (INTEGER, NOT NULL)

```python
# Write your answer for Question 2 here

query2 = """
CREATE TABLE discoveries (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    laureate_id INTEGER,
    discovery TEXT NOT NULL,
    year_discovery INTEGER NOT NULL,
    research_funding REAL NOT NULL,
    citation_count INTEGER NOT NULL,
    FOREIGN KEY (laureate_id) REFERENCES noble_laureates(id) ON DELETE CASCADE
);
"""

cursor.execute(query2)
print("Table 'discoveries' created successfully.")
```

## Data Analysis Questions

Now that you have created your database schema and inserted the data, use your tables to answer the following questions.

### Question 5: List the names of laureates who were awarded the Nobel Prize after the age of 50.

```python
# Write your query for Question 5 here
query = """
SELECT name
FROM noble_laureates
WHERE age_at_award > 50;
"""

pd.read_sql_query(query, conn)
```

### Question 6: Calculate the total prize money awarded to laureates in the field of Physiology.

```python
# Write your query for Question 6 here
query = """
SELECT SUM(prize_amount) AS total_prize_money
FROM noble_laureates
WHERE field = 'Physiology';
"""

pd.read_sql_query(query, conn)
```

### Question 7: What is the average citation count for discoveries made before 1950?

```python
# Write your query for Question 7 here

query = """
SELECT AVG(citation_count) AS avg_citations
FROM discoveries
WHERE year_discovery < 1950;
"""

pd.read_sql_query(query, conn)
```

### Question 8: Calculate the average age of laureates when they received their award, grouped by field.

```python
# Write your query for Question 8 here
query = """
SELECT field, AVG(age_at_award) AS avg_age
FROM noble_laureates
GROUP BY field;
"""

pd.read_sql_query(query, conn)
```

### Question 9: List the top three most cited discoveries along with their citation counts.

```python
# Write your query for Question 9 here
query = """
SELECT discovery, citation_count
FROM discoveries
ORDER BY citation_count DESC
LIMIT 3;
"""

pd.read_sql_query(query, conn)
```

### Question 10: Calculate the total research funding for discoveries made by laureates in each field.

```python
# Write your query for Question 10 here
query = """
SELECT nl.field, SUM(d.research_funding) AS total_funding
FROM noble_laureates nl
JOIN discoveries d ON nl.id = d.laureate_id
GROUP BY nl.field;
"""

pd.read_sql_query(query, conn)
```

### Question 11: Which laureate had the longest gap between their discovery and receiving their Nobel Prize?

```python
# Write your query for Question 11 here
query = """
SELECT nl.name,
       (nl.year_award - d.year_discovery) AS gap_years
FROM noble_laureates nl
JOIN discoveries d ON nl.id = d.laureate_id
ORDER BY gap_years DESC
LIMIT 1;
"""

pd.read_sql_query(query, conn)
```

### Question 12: Find all laureates whose names contain the string "in" (case-insensitive).

```python
# Write your query for Question 12 here
query = """
SELECT name
FROM noble_laureates
WHERE LOWER(name) LIKE '%in%';
"""

pd.read_sql_query(query, conn)
```

### Question 13: What percentage of all discoveries were funded with more than $20,000?

```python
# Write your query for Question 13 here
query = """
SELECT 
    (SUM(CASE WHEN research_funding > 20000 THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) AS percentage
FROM discoveries;
"""

pd.read_sql_query(query, conn)
```

### Question 14: Rank all discoveries by citation count within each field (Physics, Chemistry, etc.).

*Note: Requires SQLite version 3.25.0 or higher for window functions.*

```python
# Write your query for Question 14 here
query = """
SELECT 
    nl.field,
    d.discovery,
    d.citation_count,
    RANK() OVER (PARTITION BY nl.field ORDER BY d.citation_count DESC) AS rank_within_field
FROM discoveries d
JOIN noble_laureates nl ON d.laureate_id = nl.id;
"""

pd.read_sql_query(query, conn)
```

### Question 15: Calculate the average time gap (in years) between discovery and Nobel Prize award for each field.

```python
# Write your query for Question 15 here
query = """
SELECT 
    nl.field,
    AVG(nl.year_award - d.year_discovery) AS avg_gap_years
FROM noble_laureates nl
JOIN discoveries d ON nl.id = d.laureate_id
GROUP BY nl.field;
"""

pd.read_sql_query(query, conn)
```

### Question 16: Find all laureates who received their Nobel Prize more than 30 years after their discovery

```python
# Write your query for Question 16 here
query = """
SELECT nl.name
FROM noble_laureates nl
JOIN discoveries d ON nl.id = d.laureate_id
WHERE (nl.year_award - d.year_discovery) > 30;
"""

pd.read_sql_query(query, conn)
```

### Question 17: Write a query that joins both tables and shows: laureate name, field, discovery, years between discovery and award, and categorises the time gap as 'Short' (≤20 years), 'Medium' (21-40 years), or 'Long' (>40 years).

```python
# Write your query for Question 17 here
query = """
SELECT 
    nl.name,
    nl.field,
    d.discovery,
    (nl.year_award - d.year_discovery) AS gap_years,
    CASE 
        WHEN (nl.year_award - d.year_discovery) <= 20 THEN 'Short'
        WHEN (nl.year_award - d.year_discovery) BETWEEN 21 AND 40 THEN 'Medium'
        ELSE 'Long'
    END AS gap_category
FROM noble_laureates nl
JOIN discoveries d ON nl.id = d.laureate_id;
"""

pd.read_sql_query(query, conn)
```
